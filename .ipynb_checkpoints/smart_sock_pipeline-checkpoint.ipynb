```python
import os

files = maf.labeled_csv_paths(maf.DATA_DIR)
print(f'Found {len(files)} dual labeled CSV(s)\n')
for p in files[:5]:
    print('  ', os.path.basename(p))
if len(files) > 5:
    print('   …')

raw, labels, subj, mode = maf.load_csv_files(maf.DATA_DIR, labeled_only=True, raw_adc=True)
print(f'\nTotal rows: {len(labels)}   mode = {mode}')
from collections import Counter
ctr = Counter(labels)
for lab, n in ctr.most_common():
    print(f'  {lab:25s}  {n}')
```

```python
SENSOR_MAX = 4095.0

raw_all_full = raw
print(f'Concatenated raw shape =  {raw_all_full.shape}  '
      f'(i.e. {raw_all_full.shape[0]} rows ~ {raw_all_full.shape[0]/maf.SAMPLE_HZ:.1f} s '
      f' rows from  {len(files)}  CSV file(s))')

normed_all = np.clip((SENSOR_MAX - raw_all_full) / SENSOR_MAX, 0.0, 1.0)

fig, axes = plt.subplots(2, 4, figsize=(12, 4.2), sharey=True)
for ch in range(6):
    ax = axes[ch // 4, ch % 4]
    ax.hist(normed_all[:, ch], bins=60, color='#4a90e2', alpha=0.85)
    ax.set_title(f'{apre.CHANNEL_NAMES_DUAL[ch]}\n'
                 f'mean={normed_all[:, ch].mean():.2f}  '
                 f'std={normed_all[:, ch].std():.2f}', fontsize=8)
    ax.set_xlim(0, 1)
fig.suptitle(
    f'Naive norm (4095 - raw) / 4095 on all  {len(files)}  labeled CSVs concatenated)',
    fontsize=11,
)
plt.tight_layout(); plt.show()

print('\nPer-channel normed quantiles (full dataset):')
print(f'  {"channel":<12}  p01    p50    p99   mean   std')
for ch in range(6):
    col = normed_all[:, ch]
    p1, p50, p99 = np.percentile(col, [1, 50, 99])
    print(f'  {apre.CHANNEL_NAMES_DUAL[ch]:<12}  {p1:.3f}  {p50:.3f}  {p99:.3f}  '
          f'{col.mean():.3f}  {col.std():.3f}')

print('''
After naive norm, six channels idle means span ~0.02..0.48 — 
  RF memorizes channel bias; 
  new subject shifts distribution. 

''')
```

```python
ch_idx = 2  # L_Heel
long_raw = raw_all_full[:, ch_idx]
print(f'L_Heel rows =  {long_raw.shape[0]}  '
      f'(from {len(files)} CSVs concatenated)')

pre = apre.AdaptiveSensorPreprocessor(name=apre.CHANNEL_NAMES_DUAL[ch_idx])
out_raw, out_baseline, out_ratio, out_zscore = [], [], [], []
for v in long_raw:
    snap = pre.update(float(v))
    out_raw.append(snap.raw)
    out_baseline.append(snap.baseline_raw)
    out_ratio.append(snap.relative_pressure_ratio)
    out_zscore.append(snap.adaptive_zscore)

fig, axes = plt.subplots(3, 1, figsize=(12, 7), sharex=True)
axes[0].plot(out_raw, lw=0.6, label='raw')
axes[0].plot(out_baseline, lw=0.8, label='baseline_raw (online EWMA)', color='#ff6b35')
axes[0].legend()
axes[0].set_title(f'L_Heel · all  {len(files)}  CSVs raw vs online baseline ( {len(long_raw)} rows)')
axes[1].plot(out_ratio, color='#d7263d', lw=0.6)
axes[1].set_title('relative_pressure_ratio in [0,1]')
axes[1].set_ylim(-0.05, 1.05)
axes[2].plot(out_zscore, color='#1b998b', lw=0.6)
axes[2].set_title('adaptive_zscore')
axes[2].set_xlabel('Global frame index')
plt.tight_layout(); plt.show()

print(f'ratio min/max = {min(out_ratio):.3f}  {max(out_ratio):.3f}')
print(f'zscore min/max = {min(out_zscore):.2f}  {max(out_zscore):.2f}')
print('''

''')
```

```python
import joblib
import numpy as np
from IPython.display import Image, display, Markdown

print(f'{"branch":<18}{"N":>7}{"segs":>6}{"folds":>6}'
      f'{"mean":>9}{"std":>9}{"min":>9}{"OOF acc":>10}{"eval":>22}')
print('-' * 96)
for br, fn in mbm.BRANCH_TO_FILE.items():
    if not os.path.isfile(fn):
        print(f'{br:<18}  (model missing — run cell above)')
        continue
    m = joblib.load(fn).get('metrics', {})
    if not m:
        print(f'{br:<18}  (metrics missing — retrain)')
        continue
    ev = m.get('evaluation', 'unknown')
    if ev == 'blocked_intra_file':
        print(f'{br:<18}{m["n_samples"]:>7}{m["blocked_n_segments"]:>6}'
              f'{m["blocked_n_folds"]:>6}'
              f'{m["blocked_mean"]:>9.4f}{m["blocked_std"]:>9.4f}'
              f'{m["blocked_min"]:>9.4f}{m["oof_accuracy"]:>10.4f}'
              f'{ev:>22}')
    elif ev == 'leave_one_file_out':
        print(f'{br:<18}{m["n_samples"]:>7}{"-":>6}{m["lofo_n_files"]:>6}'
              f'{m["lofo_mean"]:>9.4f}{m["lofo_std"]:>9.4f}'
              f'{m["lofo_min"]:>9.4f}{m["oof_accuracy"]:>10.4f}'
              f'{ev:>22}')
    elif 'val_accuracy' in m:
        print(f'{br:<18}{m["n_samples"]:>7}  (legacy random split)  '
              f'val={m["val_accuracy"]:.4f}  test={m["test_accuracy"]:.4f}')
    else:
        print(f'{br:<18}  unknown eval mode: {ev}')

cm_dir = 'confusion_matrices'
for br in ('stairs', 'sitting', 'walking', 'standing'):
    display(Markdown(f'---\n### Branch = `{br}`'))
    fn = mbm.BRANCH_TO_FILE.get(br)
    if not fn or not os.path.isfile(fn):
        print(f'   missing model: {br}')
        continue
    m = joblib.load(fn).get('metrics', {})
    cm = m.get('confusion_oof')
    classes = m.get('classes')
    if cm is not None and classes:
        cm = np.asarray(cm, dtype=int)
        col_w = max(max(len(c) for c in classes), 6) + 2
        header = ' ' * (col_w) + ''.join(f'{c:>{col_w}}' for c in classes)
        print(header)
        for i, c in enumerate(classes):
            row = f'{c:<{col_w}}' + ''.join(f'{cm[i, j]:>{col_w}d}' for j in range(len(classes)))
            print(row + f'   n={cm[i].sum()}')
        print()
        print(f'{"class":<{col_w}}{"recall":>10}{"precision":>12}')
        for i, c in enumerate(classes):
            row_sum = cm[i].sum()
            col_sum = cm[:, i].sum()
            rec = cm[i, i] / row_sum if row_sum else float('nan')
            pre = cm[i, i] / col_sum if col_sum else float('nan')
            print(f'{c:<{col_w}}{rec:>10.4f}{pre:>12.4f}')
    for suffix in ('blocked_oof', 'lofo_oof', 'val', 'test'):
        p = os.path.join(cm_dir, f'confusion_{br}_{suffix}.png')
        if os.path.isfile(p):
            display(Image(filename=p, width=560))
            break
    else:
        print(f'   (no PNG — retrain to generate)')
```

```python
from collections import Counter

rec = rrz.OnlineRecognizer(calibration='personal_calibration.json')
assert rec._calibration.has_global_stats, 'calibration missing global stats!'
print(f'Recognizer bank frozen: {all(c._freeze for c in rec._adapt_bank.channels)}')

timeline = {k: [] for k in ['knee_mode', 'layer2', 'mode_det', 'branch_rf_key', 'state', 'gt']}
for i, (lf, lh, lk, rf, rh, rk, lab) in enumerate(rows):
    out = rec.update_bilateral((lf, lh, lk), (rf, rh, rk), t=i / maf.SAMPLE_HZ)
    timeline['knee_mode'].append(out['debug']['knee_mode'])
    timeline['layer2'].append(out['debug']['layer2_subbranch'])
    timeline['mode_det'].append(out['debug'].get('mode_detector_state', '—'))
    timeline['branch_rf_key'].append(out['debug']['branch_rf_key'])
    timeline['state'].append(out['state'])
    timeline['gt'].append(lab)

print(f'Total rows:  {len(timeline["state"])}')
print('\nFinal predicted state dist:')
for s, n in Counter(timeline['state']).most_common():
    print(f'  {s:25s}  {n}')
print('\nGround-truth (Label) dist:')
for s, n in Counter(timeline['gt']).most_common():
    print(f'  {s:25s}  {n}')
```

```python
def to_int(seq, vocab):
    return [vocab.get(s, -1) for s in seq]
v_mode = {s: i for i, s in enumerate(sorted(set(timeline['mode_det'])))}
v2 = {'MOTION_BRANCH': 1, 'STATIC_BRANCH': 0, 'knee_static_bent': 0, 'knee_dynamic_active': 1}
knee_vocab = {s: i for i, s in enumerate(sorted(set(timeline['knee_mode'])))}
states_vocab = {s: i for i, s in enumerate(sorted(set(timeline['state'])))}
gt_vocab = {s: i for i, s in enumerate(sorted(set(timeline['gt'])))}

def layer2_n(x):
    return v2.get(x, 0)

n = len(timeline['state'])
fig, axes = plt.subplots(5, 1, figsize=(12, 9), sharex=True)
axes[0].step(range(n), to_int(timeline['knee_mode'], knee_vocab), where='post', color='#457b9d')
axes[0].set_yticks(list(knee_vocab.values())); axes[0].set_yticklabels(list(knee_vocab.keys()), fontsize=6)
axes[0].set_title('Knee mode (soft-routed)')
axes[1].step(range(n), to_int(timeline['mode_det'], v_mode), where='post', color='#6a4c93')
axes[1].set_yticks(list(v_mode.values())); axes[1].set_yticklabels(list(v_mode.keys()), fontsize=7)
axes[1].set_title('mode_detector_state (MOTION/STATIC)')
axes[2].step(range(n), [layer2_n(x) for x in timeline['layer2']], where='post', color='#e07856')
axes[2].set_yticks([0, 1]); axes[2].set_yticklabels(['STATIC/L2b', 'MOTION or dyn'])
axes[2].set_title('Layer2 (motion/static branch)')
axes[3].step(range(n), to_int(timeline['state'], states_vocab), where='post', color='#2a9d8f')
axes[3].set_yticks(list(states_vocab.values())); axes[3].set_yticklabels(list(states_vocab.keys()), fontsize=7)
axes[3].set_title('Final state (predicted)')
axes[4].step(range(n), to_int(timeline['gt'], gt_vocab), where='post', color='#888')
axes[4].set_yticks(list(gt_vocab.values())); axes[4].set_yticklabels(list(gt_vocab.keys()), fontsize=7)
axes[4].set_title('Ground-truth Label in CSV')
axes[4].set_xlabel('Sample index'); plt.tight_layout(); plt.show()
```
